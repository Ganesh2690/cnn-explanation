```python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy
import numpy as np

# Create a random example dataset (replace with your own data)
data = np.random.rand(10, 10)

# Create a dendrogram for both rows and columns
row_linkage = hierarchy.linkage(data, method='average')
col_linkage = hierarchy.linkage(data.T, method='average')

# Define the row and column clustering order
row_order = hierarchy.dendrogram(row_linkage, no_plot=True)['leaves']
col_order = hierarchy.dendrogram(col_linkage, no_plot=True)['leaves']

# Create the clustered heatmap
sns.heatmap(data[row_order][:, col_order], cmap='coolwarm')

# Show the plot
plt.show()
```
